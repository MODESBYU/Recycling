This report presents the spectroscopy data collected and absorbance calculations for various deep eutectic solvents (DES). Each deep eutectic solvent was provided by the Colorado School of Mines.  
### Methods
A heated transmission cell was used to gather data. The cell was maintained at 80 C during the measurements. The solvents were first heated in a glovebox, then introduced to the cell via syringes (Fig. 1), and intensity data were taken. Before measuring each solvent, a dark measurement was taken by covering the spectrometer receptor, a baseline measurement was also taken, through the empty cell that was heated to 80 C. Python code was used to calculate the measurements, using the following equation

<center>$Absorption = -\ln({\frac{I-I_{background}}{I_0-I_{background}}})$</center>
